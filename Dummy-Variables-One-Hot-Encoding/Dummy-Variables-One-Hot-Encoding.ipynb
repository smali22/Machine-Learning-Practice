```python
import pandas as pd
house = pd.read_csv('homeprices.csv')
house
```

```python
dummy = pd.get_dummies(house.town)
dummy
```

```python
dummy = dummy.replace({True:1, False:0})
dummy
```

```python
merge = pd.concat([house,dummy], axis='columns')
merge
```

```python
final = merge.drop(['town','west windsor'],axis='columns')
final
```

```python
X = final.drop(['price'],axis='columns')
X
```

```python
y = final.price
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

### Checking Model Score

```python
model.score(X,y)
```
