```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
canada = pd.read_csv('canada_per_capita_income (1).csv')
```

```python
canada.rename(columns = {'per capita income (US$)':'capita'},inplace =True)
```

```python
canada.head()
```

```python
reg = linear_model.LinearRegression()
reg.fit(canada[['year']],canada.capita)
```

```python
plt.xlabel('Year')
plt.ylabel('Canada per capita income (US$)')
plt.scatter(canada.year,canada.capita,marker='+',color='red')
plt.show()
```

```python
reg.predict([[2020]])
```
