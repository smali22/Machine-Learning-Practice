```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv('datfa.txt')
```

```python
data.head(3)
```

```python
plt.xlabel('Area per sq ft.')
plt.ylabel('Price US$')
plt.scatter(data.area,data.price,color='red',marker='+')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(data[['area']], data.price)
```

```python
reg.predict([[5000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
